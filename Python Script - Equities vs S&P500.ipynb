```python
# 1) Collecting Data

#Installing necessary libraries:
!pip install pandas matplotlib numpy seaborn yfinance alpha_vantage quandl

#Importing libraries:

import pandas as pd                                           #For Data Manipulation/Analysis
import matplotlib.pyplot as plt                               #For Data Visualization
import numpy as np                                            #For Numerical Applicaiton
import yfinance as yf                                         #Data From Yahoo Finance

#Assigning start/end dates

start_date = '2022-01-01'
end_date = '2022-12-31'

#Choosing Assets:

# 1.Bitcoin (BTC-USD)
# 2.Coffee Jul 23 (KC=F)
# 3.EUR/CAD (EURCAD=X)
# 4.Vanguard Total Bond Market ETF (BND)
# 5.Gold ('GC=F')


#Download and printing data for Bitcoin
bitcoin_data = yf.download('BTC-USD', start = start_date, end = end_date )
print(bitcoin_data.head())
print("Total count for Bitcoin:", len(bitcoin_data))


#Download and printing data for Coffee
coffee_data = yf.download('KC=F', start = start_date, end = end_date )
print(coffee_data.head())
print("Total count for Coffee:", len(coffee_data))


#Download and printing data for EurCad
eurcad_data = yf.download('EURCAD=X', start = start_date, end = end_date )
print(eurcad_data.head())
print("Total count for eurcad:", len(eurcad_data))


#Download and printing data for Bonds
bond_data = yf.download('BND', start = start_date, end = end_date )
print(bond_data.head())
print("Total count for Bond:", len(bond_data))


#Download and printing data for Gold
gold_data = yf.download('GC=F', start = start_date, end = end_date )
print(gold_data.head())
print("Total count for Gold:", len(gold_data))

#Importing daily price for S&P 500
sp500_data = yf.download('^GSPC', start = start_date, end = end_date)
print(sp500_data.head())
print("Total count for S&P 500:", len(sp500_data))
```

```python
# 2) Processing Data

#Calculating daily percentage change for assets

bitcoin_pct_change = bitcoin_data['Adj Close'].pct_change()
coffee_pct_change = coffee_data['Adj Close'].pct_change()
eurcad_pct_change = eurcad_data['Adj Close'].pct_change()
bond_pct_change = bond_data['Adj Close'].pct_change()
gold_pct_change = gold_data['Adj Close'].pct_change()
sp500_pct_change = sp500_data['Adj Close'].pct_change()

print("Bitcoin Daily % Change:\n", bitcoin_pct_change.head(), "\nTotal count:", len(bitcoin_pct_change),"\n")
print("Coffee Daily % Change:\n", coffee_pct_change.head(), "\nTotal count:", len(coffee_pct_change),"\n")
print("EUR/CAD Daily % Change:\n", eurcad_pct_change.head(), "\nTotal count:", len(eurcad_pct_change),"\n")
print("Bond Daily % Change:\n", bond_pct_change.head(), "\nTotal count:", len(bond_pct_change),"\n")
print("Gold Daily % Change:\n", gold_pct_change.head(), "\nTotal count:", len(gold_pct_change),"\n")
print("S&P500 Daily % Change:\n", sp500_pct_change.head(), "\nTotal count:", len(sp500_pct_change),"\n")

# Plotting daily percentage change for each asset
plt.figure(figsize=(10,6))
plt.bar(bitcoin_pct_change.index, bitcoin_pct_change, color='orange',width=3.0)
plt.title('Bitcoin Daily % Change')
plt.xlabel('Date')
plt.ylabel('% Change')
plt.show()

plt.figure(figsize=(12,6))
plt.bar(coffee_pct_change.index, coffee_pct_change, color='brown',width=3.0)
plt.title('Coffee Daily % Change')
plt.xlabel('Date')
plt.ylabel('% Change')
plt.show()

plt.figure(figsize=(10,6))
plt.bar(eurcad_pct_change.index, eurcad_pct_change, color='blue',width=3.0)
plt.title('EUR/CAD Daily % Change')
plt.xlabel('Date')
plt.ylabel('% Change')
plt.show()

plt.figure(figsize=(10,6))
plt.bar(bond_pct_change.index, bond_pct_change, color='green',width=3.0)
plt.title('Bonds Daily % Change')
plt.xlabel('Date')
plt.ylabel('% Change')
plt.show()

plt.figure(figsize=(10,6))
plt.bar(gold_pct_change.index, gold_pct_change, color='gold',width=3.0)
plt.title('Gold Daily % Change')
plt.xlabel('Date')
plt.ylabel('% Change')
plt.show()

plt.figure(figsize=(10,6))
plt.bar(sp500_pct_change.index, sp500_pct_change, color='purple',width=3.0)
plt.title('S&P 500 Daily % Change')
plt.xlabel('Date')
plt.ylabel('% Change')
plt.show()

# Calculating mean, standard deviation, variance, min, and max for each data set
print("Bitcoin Stats:\n", bitcoin_pct_change.describe())
print("Coffee Stats:\n", coffee_pct_change.describe())
print("EUR/CAD Stats:\n", eurcad_pct_change.describe())
print("Bonds Stats:\n", bond_pct_change.describe())
print("Gold Stats:\n", gold_pct_change.describe())
print("S&P 500 Stats:\n", sp500_pct_change.describe())
```

```python
# 3) Regression Analysis

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error


# List of tickers
tickers = ['BTC-USD', 'KC=F', 'EURCAD=X', 'BND', 'GC=F', '^GSPC']

# Downloading data from yahoo finance
data = yf.download(tickers, start='2021-01-01', end='2022-05-15')['Adj Close'].pct_change().dropna()

# Splitting data into X and y
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

# Creating a list to store the results
results = []

# Looping through each asset
for i in range(len(tickers)-1):
    # Creating the linear regression model
    model = LinearRegression()
    # Fitting the model
    model.fit(X.iloc[:, i:i+1], y)
    # Predicting the y values
    y_pred = model.predict(X.iloc[:, i:i+1])
    # Plotting the samples and the linear model
    plt.scatter(X.iloc[:, i:i+1], y)
    plt.plot(X.iloc[:, i:i+1], y_pred, color='red')
    plt.xlabel(tickers[i])
    plt.ylabel('^GSPC')
    plt.show()
    # Storing the results
    results.append({'Asset': tickers[i], 
                    'Intercept': model.intercept_, 
                    'Coefficient': model.coef_[0], 
                    'R2': r2_score(y, y_pred), 
                    'MSE': mean_squared_error(y, y_pred)})
    
# Displaying the results
print('Results:')
for r in results:
    print(r)

# Finding the most and least correlated assets
corr = data.corr()['^GSPC']
corr = corr[corr.index != '^GSPC']
most_correlated = corr.idxmax()
least_correlated = corr.idxmin()
print('\n''Most correlated asset:', most_correlated)
print('Least correlated asset:', least_correlated)
```
